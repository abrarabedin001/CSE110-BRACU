```
#Write your code here

# TODO

class Calculator:
  def __init__(self, op1, op2 , operator):
    self.op1 = op1
    self.op2 = op2
    self.operator = operator
  def calculate_button(self):
    operator = self.operator
    if operator == "+":
      return(self.op1 + self.op2)
    elif operator == "-":
      return(self.op1 - self.op2)

    elif operator == "/":
      return(self.op1 / self.op2)

    elif operator == "*":
      return(self.op1 * self.op2)
  def details(self):
    print(str(self.op1)+self.operator+str(self.op2)+"="+str(self.calculate_button()))

#Do not change the following code:
add = Calculator(10, 20, '+')
val = add.calculate_button()
print("Returned value:", val)
add.details()
print("==================")

sub = Calculator(val, 10, '-')
val = sub.calculate_button()
print("Returned value:", val)
sub.details()
print("==================")

div = Calculator(val, 5, '*')
val = div.calculate_button()
print("Returned value:", val)
div.details()
print("==================")

mult = Calculator(val, 16, '/')
val = mult.calculate_button()
print("Returned value:", val)
mult.details()
print("==================")
```

```
#Write your code here

# TODO
class Student:
  def __init__(self,name,id,department,grades):
    self.name = name
    self.id = id
    self.department = department
    self.grades = grades
  def calculate_CGPA(self):
    sum = 0
    for r in self.grades:
      sum += r*3
    cgpa = sum/(len(self.grades)*3)
    return cgpa

  def print_details(self):

    print("--------- Department: "+self.department+" ---------")
    print("Name: "+self.name+",Id: "+str(self.id))
    
    cgpa = self.calculate_CGPA()
    print("CGPA: "+str(cgpa))
    if cgpa>=3.8:
      print("Your academic standing is 'Highest Distinction'")
    elif cgpa>=3.65:
      print("Your academic standing is 'High Distinction'")
    elif cgpa >= 3.5:
      print("Your academic standing is 'Distinction'")
    elif cgpa >= 2:
      print("Your academic standing is 'Satisfactory'")
    elif cgpa >=0:
      print("You cannot graduate")

#Do not change the following code:
s1 = Student('Dora', '15995599','CSE', [4,3.7,3.7,4])
s1.calculate_CGPA()
s1.print_details()

s2 = Student('Pingu', '12312322', 'EEE', [1.7,1.3,1.3,1.3,1])
s2.calculate_CGPA()
s2.print_details()

s3 = Student('Bob', '13311331', 'CSE', [2,3,3,3.7,2.7,2.7])
s3.calculate_CGPA()
s3.print_details()
```
